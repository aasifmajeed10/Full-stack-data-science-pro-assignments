## Question 1  
**Find the value of T(2) for the recurrence relation:**  

\[ T(n) = 3T(n-1) + 12n, \quad T(0) = 5 \]



### Solution  

We compute step by step:

- T(0) = 5  
- T(1) = 3T(0) + 12(1) = 3√ó5 + 12 = 27  
- T(2) = 3T(1) + 12(2) = 3√ó27 + 24 = **105**


```python

def T(n):
    if n == 0:
        return 5
    return 3*T(n-1) + 12*n

T(2)
```
